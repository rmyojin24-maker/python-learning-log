```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ----------------------------
# パラメータ
# ----------------------------
mu = 1.0      # 中心天体の重力パラメータ
dt = 0.001
steps = 10000

# ----------------------------
# シミュレーション関数
# ----------------------------
def simulate_basic_flyby(mu, x, y, vx, vy, dt, steps):
    x_list = []
    y_list = []
    vx_list = []
    vy_list = []
    v_list = []
    r_list = []

    for _ in range(steps):
        # 中心天体までの距離
        r = np.sqrt(x**2 + y**2)

        # 重力加速度
        ax = -mu * x / r**3
        ay = -mu * y / r**3

        # 速度更新
        vx += ax * dt
        vy += ay * dt

        # 位置更新
        x += vx * dt
        y += vy * dt

        # 記録
        v = np.sqrt(vx**2 + vy**2)

        x_list.append(x)
        y_list.append(y)
        vx_list.append(vx)
        vy_list.append(vy)
        v_list.append(v)
        r_list.append(r)

    return x_list, y_list, vx_list, vy_list, v_list, r_list
```

```python
# ----------------------------
# 初期条件
# ----------------------------
x0 = -3.0
y0 = -1.0
vx0 = 0.9
vy0 = 0.4
```

```python
x_list, y_list, vx_list, vy_list, v_list, r_list = simulate_basic_flyby(
    mu, x0, y0, vx0, vy0, dt, steps
)
```

```python
plt.figure(figsize=(7, 7))
plt.plot(x_list, y_list, label="spacecraft")
plt.scatter(0, 0, s=200, color="orange", label="central body")
plt.axis("equal")
plt.grid()
plt.legend()
plt.title("Step 2: Basic Flyby around a Fixed Body")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
time = np.arange(len(v_list)) * dt

plt.figure(figsize=(8, 4))
plt.plot(time, v_list)
plt.grid()
plt.title("Speed During Flyby")
plt.xlabel("Time")
plt.ylabel("Speed")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(time, r_list)
plt.grid()
plt.title("Distance from Central Body")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.show()
```

```python
# case 2（軌道を緩やかに）
x0 = -5.0
y0 = 0
vx0 = 1
vy0 = 0.3
x_list, y_list, vx_list, vy_list, v_list, r_list = simulate_basic_flyby(
    mu, x0, y0, vx0, vy0, dt, steps
)
```
