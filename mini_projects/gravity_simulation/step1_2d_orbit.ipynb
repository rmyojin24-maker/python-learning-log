```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# 1. パラメータ設定
# ----------------------------
mu = 1.0       # 重力の強さ
dt = 0.001     # 時間刻み
steps = 20000  # 繰り返し回数

# ----------------------------
# 2. 初期条件
# ----------------------------
x = 1.0
y = 0.0
vx = 0.0
vy = 1.0

# 軌道保存用
x_list = []
y_list = []

# ----------------------------
# 3. シミュレーション
# ----------------------------
for _ in range(steps):
    r = np.sqrt(x**2 + y**2)

    ax = -mu * x / r**3
    ay = -mu * y / r**3

    vx += ax * dt
    vy += ay * dt

    x += vx * dt
    y += vy * dt

    x_list.append(x)
    y_list.append(y)

# ----------------------------
# 4. 描画
# ----------------------------
plt.figure(figsize=(6, 6))
plt.plot(x_list, y_list, label="spacecraft orbit")
plt.scatter(0, 0, s=100, label="sun")
plt.axis("equal")
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Orbit Simulation")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# シミュレーション関数
# ----------------------------
def simulate_orbit(mu, x, y, vx, vy, dt, steps):
    x_list = []
    y_list = []

    for _ in range(steps):
        r = np.sqrt(x**2 + y**2)

        ax = -mu * x / r**3
        ay = -mu * y / r**3

        vx += ax * dt
        vy += ay * dt

        x += vx * dt
        y += vy * dt

        x_list.append(x)
        y_list.append(y)

    return x_list, y_list


# ----------------------------
# パラメータ（ここだけ変える）
# ----------------------------
mu = 1.0
dt = 0.001
steps = 20000

x0, y0 = 1.0, 0.0
vx0, vy0 = 0.0, 1.0  # ←ここ変える！


# ----------------------------
# 実行
# ----------------------------
x_list, y_list = simulate_orbit(mu, x0, y0, vx0, vy0, dt, steps)


# ----------------------------
# 描画
# ----------------------------
plt.figure(figsize=(6,6))
plt.plot(x_list, y_list)
plt.scatter(0,0, s=100)
plt.axis("equal")
plt.grid()
plt.show()
```

```python
for vy0 in [0.8, 1.0, 1.2]:
    x_list, y_list = simulate_orbit(mu, 1.0, 0.0, 0.0, vy0, dt, steps)
    plt.plot(x_list, y_list, label=f"vy={vy0}")

plt.legend()
plt.axis("equal")
plt.show()
```
